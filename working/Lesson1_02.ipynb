```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
T = 100.0
dt = 0.02
N = int(T/dt)+1
t = numpy.linspace(0.0, T, N)
```

```python
z0 = 100
b0 = 10
zt = 100
g = 9.81

u = numpy.array([z0, b0])

z = numpy.zeros(N)
z[0] = z0
```

```python
for n in range(1,N):
    u = u + dt*numpy.array([u[1], g*(1-u[0]/zt)])
    z[n] = u[0]
```

```python
pyplot.figure(figsize=(10,4))
pyplot.ylim(40,160)
pyplot.tick_params(axis='both', labelsize=14)
pyplot.xlabel('t', fontsize=14)
pyplot.ylabel('z', fontsize=14)
pyplot.plot(t,z, 'k-');
```
